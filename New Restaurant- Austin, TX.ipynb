# Opening a New Restaurant- Austin, TX

A local foodie turned entrepreneur is looking for a new venture. She recently read an article published by the U.S. Census Bureau, that listed Austin, TX as the fastest-growing city by population growth in 2020. With an already successful food blog, this new found information sparks her interest. She can instantly see the potential and lucrative opportunity to build her business. She would like to open her very own restaurant but needs to understand the lay of the the land. She figures understading her competition, most desired cuisine, and more about the customers would help her craft the best menu possible. This information will also be utilized to narrow down the location best suited for her restaurant.

In acting as her business consultant, I will successfully conclude what area or neighborhood would be the best location for her restaurant. In order to achieve this, I will be using Foursquare API to analyze the most popular locations and demographics of customers in Austin, TX. This data will be important to understand the local food industry, her competition, and demographic information on those that are dining out more frequently. This data analysis will help provide insights for her to start a business proposal, seek finnancial backing and aid in a successful business.
